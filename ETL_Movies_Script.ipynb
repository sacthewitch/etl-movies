```python
# Select the columns we want to work with from the main dataframe
columns = ['budget', 'genres', 'id', 'imdb_id', 'original_title', 'release_date', 'revenue', 'runtime']
subset_df = df[columns]
```

```python
# Check the dataframe
subset_df
```

```python
# Flatten the 'genres' column into a separate dataframe
genres_df = json_normalize(subset_df['genres'])

# Columns renamed 'col1', 'col2', 'col3', and 'col4' 
genres_df.columns = ['col1', 'col2', 'col3', 'col4']

# Pivots the data from a wide format to a long format resulting one column for the original column names and another column for the values
genres_df = genres_df.melt(id_vars=[], value_vars=['col1', 'col2', 'col3', 'col4'], var_name='original_column', value_name='value')


# Drop the 'genres' column from the subset dataframe
genres_df.drop(columns=['original_column'], inplace=True)

# Drop rows with null values 
genres_df = genres_df[genres_df['value'].notnull()]

# 'value' column is converted into a list of dictionaries
genres_df = pd.DataFrame(genres_df['value'].tolist(), columns=['id', 'name'])

# Remove dupplicates
genres_df = genres_df.drop_duplicates(subset='id')

genres_df
```

```python
# Define a function to create a list of genre names from the 'genres' column
def create_genre_list(row):
  genres = row['genres'] # Get the list of dictionaries in the 'genres' column
  genre_names = []  # Initialize an empty list to store the genre names
  for genre in genres: # Loop through the list of dictionaries
    genre_names.append(genre['name']) # Extract the 'name' key from each dictionary and append it to the list
  return genre_names # Return the list of genre names

# Create a new column 'genre_list' in the dataframe with the list of genre names
df['genre_list'] = df.apply(create_genre_list, axis=1)
```

```python
# Create dummy columns for each genre in the 'genre_list' column and Sum the dummy columns for each movie
genre_dummies = pd.get_dummies(df['genre_list'].apply(pd.Series).stack()).sum(level=0)

# Join the dummy columns to the subset dataframe
subset_df = subset_df.join(genre_dummies)
```

```python
# Check the dataframe
subset_df
```

```python
# Drop the 'genres' column from the subset dataframe
subset_df.drop(columns=['genres'], inplace=True)

# Copy as movie_df
movie_df = subset_df.copy()
```

```python
# Check the dataframe
movie_df
```

```python
def runtime_to_hours_minutes(runtime):
    hours = runtime // 60  # Integer division to get the number of hours
    minutes = runtime % 60  # Modulo operator to get the remaining minutes
    return f"{hours:02d}:{minutes:02d}"  # Return a string in the format "HH:MM"

    
```

```python
# Iterate over the runtime column and apply the runtime_to_hours_minutes function
for i, row in movie_df.iterrows():
    movie_df.at[i, 'runtime'] = runtime_to_hours_minutes(row['runtime'])
```

```python
movie_df
```

```python
# Convert the 'release_date' column to datetime
subset_df['release_date'] = pd.to_datetime(subset_df['release_date'])

# Extract the year, month, and day from the 'release_date' column and create separate columns for each
subset_df[['year', 'month', 'day']] = subset_df['release_date'].apply(lambda x: pd.Series([x.year, x.month, x.day]))
```

```python
# Check the dataframe
subset_df
```

```python
# Define a function to extract the day of the week from the 'release_date' column
def get_day_of_week(row):
  day_of_week = row['release_date'].weekday() # Use the dt.weekday accessor to extract the day of the week
  # Map integer values to string values
  day_of_week_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
  return day_of_week_map[day_of_week]

# Apply the 'get_day_of_week' function to the 'release_date' column and create a new 'day_of_week' column
subset_df['day_of_week'] = subset_df.apply(get_day_of_week, axis=1)
```

```python
# Check the dataframe
subset_df
```

```python
# Create a new dataframe 'expanded_release_date_df' with 
# only the 'id', 'release_date', 'day', 'month', 'year', and 'day_of_week' columns
expanded_release_date_df = subset_df[['id', 'release_date', 'day', 'month', 'year', 'day_of_week']]
```

```python
# Check the dataframe
expanded_release_date_df
```

```python
# Export to CSV files
genres_df.to_csv('genres.csv')

movie_df.to_csv('movie.csv')

expanded_release_date_df.to_csv('datetimes.csv')
```
